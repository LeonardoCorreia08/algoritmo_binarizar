```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python


def binarizar_imagem(imagem):
    largura, altura = imagem.size

    # Converter a imagem em escala de cinza
    imagem_cinza = Image.new("L", (largura, altura))
    for x in range(largura):
        for y in range(altura):
            r, g, b = imagem.getpixel((x, y))
            valor_cinza = int((r + g + b) / 3)
            imagem_cinza.putpixel((x, y), valor_cinza)

    # Converter a imagem binária
    imagem_binaria = Image.new("1", (largura, altura))
    for x in range(largura):
        for y in range(altura):
            pixel_cinza = imagem_cinza.getpixel((x, y))
            valor_binario = 255 if pixel_cinza >= 128 else 0
            imagem_binaria.putpixel((x, y), valor_binario)

    return imagem_cinza, imagem_binaria
```

```python
# Carregar a imagem
caminho_imagem = r"C:\Users\My\Downloads\Projeto 2\ThanosManopla_des.jpg"
imagem = Image.open(caminho_imagem)
```

```python
# Chamada da função para binarizar a imagem
imagem_cinza, imagem_binaria = binarizar_imagem(imagem)
```

```python
# Exibir as versões da imagem usando Matplotlib
plt.subplot(1, 2, 1)
plt.imshow(imagem_cinza, cmap="gray")
plt.title("Imagem em escala de cinza")

plt.subplot(1, 2, 2)
plt.imshow(imagem_binaria, cmap="binary")
plt.title("Imagem binária")

plt.show()
```
